```python
#read csv file and convert to txt file
import csv

#
count=0
#read csv file
with open('/data/len_8000_fold_split/split_1/train/fold_1.csv', 'r') as csvfile:
    #print the read file
    #remove header
    next(csvfile)
    a=csv.reader(csvfile)
    for row in a:
        

        if row[1] == '1':
            #create a new txt file with name as count
            

            
            with open('/data/promoter_prediction/train/positive/'+str(count)+'.txt', 'w') as f:
                f.write(row[0])
            count+=1
        elif row[1] == '0':
            #create a new txt file
            
            with open('/data/promoter_prediction/train/negative/'+str(count)+'.txt', 'w') as f:
                f.write(row[0])
            count+=1
        
        else:
            print("Error: label should be 0 or 1")
        
with open('/data/len_8000_fold_split/split_1/train/fold_2.csv', 'r') as csvfile:
    #print the read file
    #remove header
    next(csvfile)
    a=csv.reader(csvfile)
    for row in a:
        

        if row[1] == '1':
            #create a new txt file with name as count

            
            with open('data/promoter_prediction/train/positive/'+str(count)+'.txt', 'w') as f:
                f.write(row[0])
            
            count+=1
        
        elif row[1] == '0':
            #create a new txt file
            
            with open('data/promoter_prediction/train/negative/'+str(count)+'.txt', 'w') as f:
                f.write(row[0])
            count+=1
        
        else:
            print("Error: label should be 0 or 1")


with open('/data/len_8000_fold_split/split_1/train/fold_3.csv', 'r') as csvfile:
    #print the read file
    #remove header
    next(csvfile)
    a=csv.reader(csvfile)
    for row in a:
        

        if row[1] == '1':
            #create a new txt file with name as count

            
            with open('/data/promoter_prediction/train/positive/'+str(count)+'.txt', 'w') as f:
                f.write(row[0])
            
            count+=1
        elif row[1] == '0':
            #create a new txt file
            
            with open('/data/promoter_prediction/train/negative/'+str(count)+'.txt', 'w') as f:
                f.write(row[0])
            count+=1
        
        else:
            print("Error: label should be 0 or 1")
    
```
