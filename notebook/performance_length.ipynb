```python
import numpy as np
import matplotlib.pyplot as plt
from brokenaxes import brokenaxes

np.random.seed(19680801)

hyena_dna_length=[512,1000,2000,3000,30000]
hyena_dna_length_2=[512,1000,2000,3000]

hyena_dna=[0.406,0.575,0.582,0.677,0.989]
NT=[0.514,0.602,0.772,0.786]
bert2=[0.49,0.64,0.741,0.747]
genalm=[0.527,0.603,0.727,0.773]
Caduceus=[0.464,0.560,0.640,0.711,0.983]


fig = plt.figure(figsize=(8,6))
ax = brokenaxes(xlims=((400, 3030), (29970, 30100)),despine=True, wspace=0.15,d=0.01)
ax.plot(hyena_dna_length,hyena_dna,label='Hyena-DNA',marker='o',linewidth=3,markersize=10)
ax.plot(hyena_dna_length_2,NT,label='NT',marker='o',linewidth=3,markersize=10)        
ax.plot(hyena_dna_length_2,bert2,label='DNABERT-2', marker='o',linewidth=3,markersize=10)
ax.plot(hyena_dna_length_2,genalm,label='GENA-LM',marker='o',linewidth=3,markersize=10)
ax.plot(hyena_dna_length,Caduceus,label='Caduceus',marker='o',linewidth=3,markersize=10)
ax.grid(ls='--',alpha=0.5,linewidth=1)

ax.set_xlabel('Input Length',fontsize=14)
ax.set_ylabel('Accuracy',fontsize=14)

plt.title('Accuracy on Various Lengths in Species Classification Task',fontsize=14)
ax.legend(loc='upper left',fontsize=14)

plt.ylim(0.4,1.0)

plt.grid(axis='y', alpha=0.4)
plt.grid(axis='x', alpha=0.4)
# plt.show()
plt.savefig('performance_length_species.pdf',dpi=300)
```
