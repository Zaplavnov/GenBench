```python
from Bio import SeqIO
import os
input_file = 'data/plant-genomic-benchmark/poly_a/arabidopsis_thaliana_test.fa'
output_file='data/plant_genomic_benchmark_txt/poly_a/arabidopsis_thaliana/test/'
fasta_sequences = SeqIO.parse(open(input_file),'fasta')
i=0
for fasta in fasta_sequences:
        name, sequence = fasta.id, str(fasta.seq)
        #splite name using '|'

        name_list = name.split('|')
        i+=1
        #convert i to string
        if(name_list[1] == '0'):
            #make txt file if file do not exist
                
            if not os.path.exists(output_file+'0/'+str(i)+'.txt'):
                with open(output_file+'0/'+str(i)+'.txt', 'w') as f:
                    f.write(sequence)
            #append sequence to file
        elif(name_list[1] == '1'):
            #make txt file if file do not exist
                
            if not os.path.exists(output_file+'1/'+str(i)+'.txt'):
                with open(output_file+'1/'+str(i)+'.txt', 'w') as f:
                    f.write(sequence)
            #save sequence to file
        else:
            print('error')
            
                
        
```

```python
from sklearn.datasets import make_multilabel_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
import numpy as np
from sklearn.multioutput import MultiOutputClassifier
X, y = make_multilabel_classification(random_state=0)
inner_clf = LogisticRegression(solver="liblinear", random_state=0)
clf = MultiOutputClassifier(inner_clf).fit(X, y)
y_score = np.transpose([y_pred[:, 1] for y_pred in clf.predict_proba(X)])
roc_auc_score(y, y_score, average=None)
```
